根据下列递归公式, 计算下列 $T(n)$ 对应的的渐近上界。要求所求的边界尽可能的紧（tight）, 请写明步骤。

1. $T(1)=1; T(n)=T(n/4)+1$ for $n>1$
2. $T(1)=1;T(n)=3T(n/3)+n^2$ for $n>1$
3. $T(1)=1;T(n)=T(2n/3)+1$ for $n>1$
4. $T(1)=1;T(n)=5T(n/4)+n$ for $n>1$
5. $T(n)=1 \ for\ n \le 2 ; T(n)=T(\sqrt{n})+1 \ for \ n>2$

### 递归式渐近上界求解

---

#### 1. $T(n) = T(n/4) + 1$
- 方法：主定理 / 递推展开
- 展开：$T(n) = T(1) + \log_4 n$
- 渐近上界：$\Theta(\log n)$

---

#### 2. $T(n) = 3T(n/3) + n^2$
- 方法：主定理（情况3）
- $n^{\log_3 3} = n$，而 $n^2$ 多项式大于 $n$
- 渐近上界：$\Theta(n^2)$

---

#### 3. $T(n) = T(2n/3) + 1$
- 方法：主定理 / 递推展开
- 展开：$T(n) = T(1) + \log_{3/2} n$
- 渐近上界：$\Theta(\log n)$

---

#### 4. $T(n) = 5T(n/4) + n$
- 方法：主定理（情况1）
- $n^{\log_4 5} \approx n^{1.16}$，而 $n$ 多项式小于它
- 渐近上界：$\Theta(n^{\log_4 5})$

---

#### 5. $T(n) = T(\sqrt{n}) + 1$
- 方法：变量替换 $n = 2^{2^k}$
- 令 $S(k) = T(2^{2^k})$，则 $S(k) = S(k-1) + 1$，解得 $S(k) = k + C$
- 还原：$k = \log_2\log_2 n$，所以 $T(n) = \Theta(\log\log n)$
- 渐近上界：$\Theta(\log\log n)$

---

## 问题 2

观察下方的分治算法（divide-and-conquer algorithm）的伪代码, 回答下面问题

```latex
DoSomething(A,p,r)
-----
n := r-p+1
if n=2 and A[p]>A[r] then
    swap A[p] and A[r]
else if n >= 3 then

    DoSomething(A,p,p+m-1)
    DoSomething(A,r-m+1,r)

first call: DoSomething(A,1,n)
```

note：$ceil(2n/3)=\left\lceil {2n/3} \right\rceil$；$:=$ 表示赋值, 等价于 $\to$；A是一个包含n的整数元素的数组, 

1. 写出该算法时间复杂度的递归公式, 并求解其对应的渐进表示
2. 描述一下该算法的功能, 并判断是否是最高效的解决方案
3. \* 使用python编程实现上述算法或其对应的更高效版本的算法

idea:

1.递归公式为：$$T(n) =
\begin{cases}
O(1) & n \le 2 \\
3T\left(\left\lceil \frac{2n}{3} \right\rceil\right) + O(1) & n \ge 3
\end{cases}$$


渐进表示为：$$ T(n) = \Theta\left(n^{\log_{\frac{3}{2}} 3}\right) \approx \Theta\left(n^{2.71}\right) $$

2.将整个数组升序排列（当数组长度为 2 时，直接比较并交换，保证升序；当数组长度 ≥3 时，将数组分为两部分：前 ⌈2n/3⌉ 个元素、后 ⌈2n/3⌉ 个元素，递归排序前部分 → 递归排序后部分 → 再次递归排序前部分），利用 “两次前半段排序 + 一次后半段排序” 的方式，逐步将大元素 “推” 到数组末尾。
该算法的时间复杂度远高于最优排序算法的时间复杂度O(nlogn)，所以不是最高效的解决方案。

````python
# 第3问答案
# 原算法
import math

def DoSomething(A, p, r):
    n = r - p + 1
    if n == 2:
        if A[p] > A[r]:
            A[p], A[r] = A[r], A[p]
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        DoSomething(A, p, p + m - 1)
        DoSomething(A, r - m + 1, r)
        DoSomething(A, p, p + m - 1)

# 测试
if __name__ == "__main__":
    arr = [3, 1, 4, 1, 5, 9, 2, 6]
    print("第3问朴素版运行结果：")
    print("原数组:", arr)
    DoSomething(arr, 0, len(arr)-1)
    print("排序后:", arr)
    

# 更高效的版本（归并排序）
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 测试
if __name__ == "__main__":
    arr = [3, 1, 4, 1, 5, 9, 2, 6]
    print("第3问高效版运行结果：")
    print("原数组:", arr)
    sorted_arr = merge_sort(arr)
    print("归并排序后:", sorted_arr)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

````python
# 这里的答案包括第1问和第2问
"""
算法思路：
贪心算法，注意这里的profit是收益，不是价格，所以需要找的是连续子数组的最大和。

时间复杂度为O(n)
"""

def maxProfit(profits):
    result = float('-inf')
    count = 0
    start = 0
    best_start = 0
    best_end = 0

    for i in range(len(profits)):
        count += profits[i]

        # 更新最大和 & 记录区间
        if count > result:
            result = count
            best_start = start
            best_end = i

        # 和为负，清空，重新开始
        if count <= 0:
            count = 0
            start = i + 1

    return result, best_start, best_end


# 测试
if __name__ == "__main__":
    profits = [3,2,1,-7,5,2,-1,3,-1]
    profit, buy, sell = maxProfit(profits)
    
    print("第1、2问运行结果：")
    print("收益数组：", profits)
    print(f"最佳买入日期：第 {buy + 1} 天")
    print(f"最佳卖出日期：第 {sell + 1} 天")
    print(f"最大收益：{profit}")
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

````python
# 第1问答案
"""
算法思路：
二分查找，取中间位置 mid
若 A[mid] 是局部最小值（比左右邻居都小），直接返回
若 A[mid] > A[mid-1]：说明左侧一定存在局部最小值，递归 / 二分左半区间
若 A[mid] > A[mid+1]：说明右侧一定存在局部最小值，递归 / 二分右半区间

时间复杂度为O(logn)
"""

def find_one_local_min(A):
    n = len(A)
    left, right = 0, n - 1
    while left <= right:
        mid = (left + right) // 2
        # 边界处理
        if mid == 0:
            if A[mid] < A[mid + 1]:
                return A[mid]
            else:
                left = mid + 1
        elif mid == n - 1:
            if A[mid] < A[mid - 1]:
                return A[mid]
            else:
                right = mid - 1
        else:
            if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
                return A[mid]
            elif A[mid] > A[mid - 1]:
                right = mid - 1
            else:  # A[mid] > A[mid + 1]
                left = mid + 1
    return None  # 理论上不会执行到这里

# 测试
A = [9, 3, 7, 2, 1, 4, 5]
print("第1问运行结果：",find_one_local_min(A))  # 输出 3 或 1，取决于二分路径


# 第2问答案
"""
算法思路：
找出所有局部最小值，直接遍历数组即可
遍历数组中第 1 到 n-2 个元素
对每个元素，判断是否同时小于左右邻居
收集所有满足条件的元素

时间复杂度为O(n)
"""

def find_all_local_min(A):
    n = len(A)
    res = []
    for i in range(1, n - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            res.append(A[i])
    return res

# 测试
A = [9, 3, 7, 2, 1, 4, 5]
print("第2问运行结果：",find_all_local_min(A))  # 输出 [3, 1]
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

````python
# 第1问答案
"""
算法思路：
使用快速排序，排完序后，取前k个元素即可

时间复杂度O(nlogn)
"""

def k_smallest_by_sort(arr, k):
    # 复制数组避免修改原数组
    sorted_arr = sorted(arr)
    return sorted_arr[:k]

# 测试
arr = [5,4,3,2,6,1,88,33,22,107]
print("第1问运行结果：", k_smallest_by_sort(arr, 3))  # 输出: [1, 2, 3]

# 第2问答案
"""
算法思路：
使用最小堆，先将所有元素插入到最小堆中，然后进行k次Extract-Min操作，得到前k小的元素。

建堆并插入的时间为O(nlogn)，k次Extract-Min的时间为O(klogn)，总时间复杂度为O((n + k)logn)
"""

import heapq

def k_smallest_by_heap(arr, k):
    heap = []
    # 往空堆中不断插入元素
    for num in arr:
        heapq.heappush(heap, num)
    # 调用k次 Extract-Min
    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))
    return result

# 测试
arr = [5,4,3,2,6,1,88,33,22,107]
print("第2问运行结果：", k_smallest_by_heap(arr, 3))  # 输出: [1, 2, 3]

# 第3问答案
"""
算法思路：
使用大顶堆来维护当前找到的前 k 小元素，初始化一个大小为 k 的大顶堆。
遍历数据（可以边读边处理，不用全部加载到内存）：
当堆的大小 < k：直接将当前元素压入堆。
当堆的大小 == k：
如果当前元素 < 堆顶元素（堆顶是当前前 k 小里的最大值）：弹出堆顶，压入当前元素。
否则跳过。
遍历结束后，堆中就是前 k 小的元素，再对堆内元素排序即可得到升序结果。

遍历所有数据O(n)，每次堆操作O(logk)，整体时间复杂度为O(nlogk)
"""
import heapq

def k_smallest_large_data(arr, k):
    max_heap = []
    for num in arr:
        if len(max_heap) < k:
            # 用负数模拟大顶堆
            heapq.heappush(max_heap, -num)
        else:
            # 当前元素比堆顶（前k个里的最大值）小，替换
            if num < -max_heap[0]:
                heapq.heappop(max_heap)
                heapq.heappush(max_heap, -num)
    # 堆中是前k小的元素，取反后排序
    result = [-x for x in max_heap]
    result.sort()
    return result

# 测试
arr = [5,4,3,2,6,1,88,33,22,107]
print("第3问运行结果：", k_smallest_large_data(arr, 3))  # 输出: [1, 2, 3]
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

````python
# 答案
"""
算法思路：
使用栈来模拟星球的碰撞过程，遍历每一个星球，依次压入栈中。
每次压入前，检查栈顶的星球和当前星球是否会发生碰撞。
    栈顶是向右（正数），当前星球是向左（负数），会碰撞。
    其他情况直接入栈。
碰撞时，
    如果 abs(当前星球) > abs(栈顶星球)：栈顶星球消失（弹出），继续检查新的栈顶。
    如果 abs(当前星球) < abs(栈顶星球)：当前星球消失，不入栈，结束检查。
    如果质量相等：当前星球消失，栈顶保留，结束检查（只把当前星球标记为死亡）。

每个星球最多入栈一次、出栈一次，因此总的操作次数为 O(n)，时间复杂度为 O(n)
"""

def planet_collision(A):
    stack = []
    for planet in A:
        alive = True
        while alive and stack and stack[-1] > 0 and planet < 0:
            top = stack[-1]
            if abs(planet) > abs(top):
                stack.pop()
            elif abs(planet) < abs(top):
                alive = False
            else:
                # 质量相等，当前星球消失，栈顶保留（只标记当前星球死亡，不弹出栈顶）
                # stack.pop()
                alive = False
        if alive:
            stack.append(planet)
    return stack

# 测试
if __name__ == "__main__":
    # 测试例1
    A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
    print(planet_collision(A1))
    # 测试例2
    A2 = [23, -8, 9, -3, -7, 9, -23, 22]
    print(planet_collision(A2))
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

````python
# 答案
"""
算法思路：
动态规划，n 是大理石的个数，max_weight 是背包的最大容量。
dp[i][j] 表示从下标为[0-i]的物品里任意取，放进容量为j的背包，价值总和最大是多少。
状态转移方程：
dp[i][j] = max(dp[i-1][j], dp[i-1][j-weight[i]] + value[i])

时间复杂度为O(n * max_weight)
"""
def knapsack_max_value(items, max_weight):
    n = len(items)
    # 初始化 dp
    dp = [[0] * (max_weight + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        weight, value = items[i-1]
        for j in range(max_weight + 1):
            if j < weight:
                # 装不下，只能不选
                dp[i][j] = dp[i-1][j]
            else:
                # 选或不选，取最大
                dp[i][j] = max(dp[i-1][j], dp[i-1][j - weight] + value)
    
    return dp[n][max_weight]

# 测试
items = [(5,10), (4,40), (6,30), (3,50)]
max_weight = 9
print("最大价值为：", knapsack_max_value(items, max_weight))
````

````python
# 答案
"""
算法思路：
中序遍历，二叉搜索树中序遍历一定是递增序列，所以先把所有节点按中序存到一个列表里。
遍历这个列表，找破坏递增顺序的地方（逆序对）
    第一次出现 pre.val > cur.val：pre 是错误节点 x
    第二次 / 最后一次出现：cur 是错误节点 y
交换 x 和 y 的值，树就恢复成正常二叉搜索树了。

时间复杂度是O(n)
"""

def recoverTree(self, root):
    nodes = []
    
    # 中序遍历，把所有节点按顺序存到列表里
    def dfs(root):
        if not root:
            return
        dfs(root.left)
        nodes.append(root)  # 存节点，不是存值
        dfs(root.right)
    
    dfs(root)
    
    x = None  # 第一个错误节点（大的那个）
    y = None  # 第二个错误节点（小的那个）
    pre = nodes[0]
    
    # 遍历有序列表，找逆序对
    for i in range(1, len(nodes)):
        if pre.val > nodes[i].val:
            y = nodes[i]
            if not x:
                x = pre
        pre = nodes[i]
    
    # 交换两个错误节点的值
    if x and y:
        x.val, y.val = y.val, x.val

# 注：该题没有直接打印输出，但在leetcode上已验证通过。
````

## 问题 9 \*
给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。



<div style="text-align:center;width:800px;"> <img alt="" src="./fig/hw3q6.png"> </div>

请分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

````python
# 答案
"""
算法思路：
双指针，存水量 = min (左边最高，右边最高) - 当前柱子高度。
用两个数组分别预处理：
    maxLeft[i]：第i个位置左边的最高柱子高度
    maxRight[i]：第i个位置右边的最高柱子高度
遍历数组，计算每个位置的接雨水量，并累加。

时间复杂度为O(n)
"""

def trap(height):
    if len(height) <= 2:
        return 0
    
    n = len(height)
    maxLeft = [0] * n # 每个位置左边最高柱子高度
    maxRight = [0] * n # 每个位置右边最高柱子高度
    
    # 初始化左边最大值
    maxLeft[0] = height[0]
    for i in range(1, n):
        maxLeft[i] = max(height[i], maxLeft[i - 1])
    
    # 初始化右边最大值
    maxRight[-1] = height[-1]
    for i in range(n - 2, -1, -1):
        maxRight[i] = max(height[i], maxRight[i + 1])
    
    # 计算接雨水总量
    total = 0
    for i in range(n):
        water = min(maxLeft[i], maxRight[i]) - height[i]
        if water > 0:
            total += water
    return total

# 测试
if __name__ == "__main__":
    # 示例1
    print("示例1：")
    height1 = [0,1,0,2,1,0,1,3,2,1,2,1]
    print("输入：", height1)
    print("输出：", trap(height1))
    
    # 示例2
    print("\n示例2：")
    height2 = [4,2,0,3,2,5]
    print("输入：", height2)
    print("输出：", trap(height2))
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:

````python
# 答案
"""
算法思路：
使用深度优先搜索（DFS），从根节点到叶子节点的每一条路径都代表一个数字。
初始化 x = 0，把 x 作为 DFS 的参数，并在 DFS 中按照 x = x * 10 + node.val去更新x
遇到叶子节点就返回当前数字，最后把所有路径结果相加。

时间复杂度为O(n)
"""

# 定义树节点
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def sumNumbers(self, root: TreeNode, x=0) -> int:
    # 空节点返回 0
    if not root:
        return 0

    # 拼接当前路径数字
    x = x * 10 + root.val

    # 叶子节点，返回当前路径的数字
    if not root.left and not root.right:
        return x

    # 左 + 右 递归求和
    return self.sumNumbers(root.left, x) + self.sumNumbers(root.right, x)

# 注：该题没有直接打印输出，但在leetcode上已验证通过。
````
